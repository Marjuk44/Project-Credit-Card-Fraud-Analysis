```python
import pandas as pd
import sqlite3

df = pd.read_csv("creditcard.csv")

conn = sqlite3.connect('creditcard_fraud.db')
cursor = conn.cursor()

# Create the Transactions table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Transactions (
        Time FLOAT,
        V1 FLOAT,
        V2 FLOAT,
        V3 FLOAT,
        V4 FLOAT,
        V5 FLOAT,
        V6 FLOAT,
        V7 FLOAT,
        V8 FLOAT,
        V9 FLOAT,
        V10 FLOAT,
        V11 FLOAT,
        V12 FLOAT,
        V13 FLOAT,
        V14 FLOAT,
        V15 FLOAT,
        V16 FLOAT,
        V17 FLOAT,
        V18 FLOAT,
        V19 FLOAT,
        V20 FLOAT,
        V21 FLOAT,
        V22 FLOAT,
        V23 FLOAT,
        V24 FLOAT,
        V25 FLOAT,
        V26 FLOAT,
        V27 FLOAT,
        V28 FLOAT,
        Amount FLOAT,
        Class INTEGER
    )
''')
conn.commit()

df.to_sql('Transactions', conn, if_exists='replace', index=False)

conn.close()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

conn = sqlite3.connect('creditcard_fraud.db')

df = pd.read_sql_query('SELECT * FROM Transactions', conn)

conn.close()

df['Time'] = pd.to_timedelta(df['Time'], unit='s')

# Create a new column for hour
df['Hour'] = (df['Time'].dt.seconds // 3600) % 24

# Set 'Time' as index for resampling
df.set_index('Time', inplace=True)

# Resample the data to hourly intervals and count fraudulent transactions
hourly_fraud = df[df['Class'] == 1].resample('H').size()

# Count the number of fraudulent transactions per hour for heatmap
hourly_fraud_counts = df[df['Class'] == 1].groupby('Hour').size()
hourly_fraud_counts = hourly_fraud_counts.reindex(range(24), fill_value=0)

# Plotting
plt.figure(figsize=(18, 12))

# Histogram
plt.subplot(3, 1, 1)
sns.histplot(df[df['Class'] == 1]['Hour'], bins=24, kde=False)
plt.title('Fraudulent Transactions Over Time')
plt.xlabel('Time (Hours)')
plt.ylabel('Number of Fraudulent Transactions')
plt.xticks(range(0, 25))

# Line Plot
plt.subplot(3, 1, 2)
plt.plot(hourly_fraud.index, hourly_fraud.values, marker='o')
plt.title('Hourly Fraudulent Transactions')
plt.xlabel('Time (Hours)')
plt.ylabel('Number of Fraudulent Transactions')
plt.xticks(range(0, 25))
plt.grid(True)

# Heatmap
plt.subplot(3, 1, 3)
sns.heatmap(hourly_fraud_counts.values.reshape(1, -1), cmap='Reds', cbar=True, xticklabels=range(24), yticklabels=[])
plt.title('Heatmap of Fraudulent Transactions by Hour')
plt.xlabel('Hour of Day')

# Show plots
plt.tight_layout()
plt.show()
```

```python
conn = sqlite3.connect('creditcard_fraud.db')
cursor = conn.cursor()

# Total number of transactions
cursor.execute('SELECT COUNT(*) FROM Transactions')
total_transactions = cursor.fetchone()[0]

# Total number of fraudulent transactions
cursor.execute('SELECT COUNT(*) FROM Transactions WHERE Class = 1')
total_fraudulent = cursor.fetchone()[0]

# Average transaction amount
cursor.execute('SELECT AVG(Amount) FROM Transactions')
avg_amount = cursor.fetchone()[0]

conn.close()

print(f"Total Transactions: {total_transactions}")
print(f"Total Fraudulent Transactions: {total_fraudulent}")
print(f"Average Transaction Amount: {avg_amount:.2f}")
```

```python
import sqlite3

# Reopen the connection
conn = sqlite3.connect('creditcard_fraud.db')
cursor = conn.cursor()

# Example: Query to get the count of fraudulent transactions by hour
query = '''
    SELECT Hour, COUNT(*) as Fraud_Count
    FROM (
        SELECT (CAST(Time AS INTEGER) / 3600) % 24 AS Hour, Class
        FROM Transactions
        WHERE Class = 1
    )
    GROUP BY Hour
    ORDER BY Fraud_Count DESC
'''

cursor.execute(query)
hourly_fraud_data = cursor.fetchall()

# Close the connection after fetching the data
conn.close()

# Display the results
for hour, count in hourly_fraud_data:
    print(f"Hour: {hour}, Fraud Count: {count}")
```

```python
conn = sqlite3.connect('creditcard_fraud.db')
cursor = conn.cursor()

query = '''
    SELECT Class, COUNT(*) as Transaction_Count
    FROM Transactions
    GROUP BY Class
'''

cursor.execute(query)
class_transaction_data = cursor.fetchall()

conn.close()

for class_type, count in class_transaction_data:
    print(f"Class: {class_type}, Transaction Count: {count}")

```

```python
conn = sqlite3.connect('creditcard_fraud.db')
cursor = conn.cursor()

query = '''
    SELECT Class, AVG(Amount) as Average_Amount
    FROM Transactions
    GROUP BY Class
'''

cursor.execute(query)
class_average_amount = cursor.fetchall()

conn.close()

for class_type, avg_amount in class_average_amount:
    print(f"Class: {class_type}, Average Amount: {avg_amount:.2f}")
```
